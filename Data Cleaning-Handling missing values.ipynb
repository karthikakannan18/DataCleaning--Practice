```python
#importing the libraries
import pandas as pd
import numpy as py
```

```python
#reading the data
data = pd.read_csv('NFL 2009-2016.csv')
```

```python
#set seed for reproducibility
py.random.seed(0)
```

```python
#looking at the first ten rows of the NFL data
data.head(10)
```

```python
# to find the number of missing data points 
#getting the number if missing data points per column

missing_value_count = data.isnull().sum()

#looking at the # of missing points in the first twenty columns

missing_value_count[0:20]
```

```python
#to find how many total missing values in the dataset

total_cells = py.product(data.shape)
total_missing = missing_value_count.sum()

#percent of missing data 

percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
#looking at the # of missing points in the first twenty columns

missing_value_count[0:20]
```

```python
#removing all the rows that contain a missing value

data.dropna()
```

```python
#removing all columns with at least one missing value

columns_with_na_dropped = data.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
#to find how much data did we lose

print("Columns in original dataset: %d \n" % data.shape[1])
print("Columns in na's dropped dataset: %d \n" % columns_with_na_dropped.shape[1])
```

```python
# Filling missing values automatically

subset_data = data.loc[:, 'EPA':'Season'].head()
subset_data
```

```python
#replace all NA's with 0 
subset_data.fillna(0)
```

```python
# replacing all NA's the value that comes directly after it in the same column
#then replace all the remainging na's with 0 

subset_data.fillna(method='bfill', axis=0).fillna(0)
```
